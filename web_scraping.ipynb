```
from bs4 import BeautifulSoup
```

```
import requests
```

```
url='https://en.wikipedia.org/wiki/Firewall_(computing)'
```

```
# we are sending request to url and taling the response
response=requests.get(url)
```

```
response
```

```
response.text
```

```
# parsing library - html.parser and second is lxml
soup=BeautifulSoup(response.text, "lxml")
```

```
soup
```

```
for Alink in soup.find_all('a'):
  print(Alink)
```

```
for Alink in soup.find_all('a'):
   print(Alink)
```

```
for ptag in soup.find_all('p'):
  #print(ptag)
  # now second loop for b tab
   for btag in ptag.find_all('b'):
    print(btag.text)
```
